# Converting iris data into 3 dimensions and plotting a 3D graph 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
y.shape
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(X)
```

```python
pca.components_
```

```python
Z = pca.transform(X)
```

```python
Z.shape
```

```python
#now the dataset is transformed into 3D so now we need to plot the dataset
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(Z[:,0] , Z[:,1] , Z[:,2] , c=y)
```
